```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sumbit=pd.read_csv('data/sample_submit.csv',header=None)
sumbit.columns=['id','delay_time']
```

```python
df_test=pd.read_csv('data/test.csv')
df_test['delayTime']=df_test['delayTime'].fillna(-1)
```

```python
delayTime=[]
for name,data in df_test.groupby('trainNo'):
    before_delay=0
    delay=0
    for _,rows in data.iterrows():
        if rows['delayTime']!=-1 and rows['target']==0:
            delay=rows['delayTime']
        elif rows['target']==1:
            delayTime.append(delay)
```

```python
sumbit['delay_time']=delayTime
```

```python
sumbit.to_csv('data/sumbit.csv',header=None,index=None)
```

```python
plt.subplots(figsize=(20,8))
sumbit['delay_time'].value_counts().plot(kind='bar')
```
